## Question 1

Identify and list all unique customer IDs who have made returns between July 1st 2024 and June 30th 2025. This will help us understand the base set of customers involved in returns during the specified period.

```python
# Group returned customers
returned_customers = customer_returns[
    (customer_returns['return_flag'] == True) &
    (customer_returns['order_date'] >= '2024-07-01') &
    (customer_returns['order_date'] <= '2025-06-30')
]['customer_id'].drop_duplicates()

print(returned_customers.tolist())
```

## Question 2

Convert the 'order_date' column to a datetime format and create a MultiIndex with 'customer_id' and 'order_date'. Then, calculate the total number of returns per customer for each month. This will provide insights into monthly return patterns for each customer.

```python
# Convert to datetime, allowing mixed formats
customer_returns['order_date'] = pd.to_datetime(
    customer_returns['order_date'],
    format='mixed',  # lets Pandas handle both dash and slash formats
    errors='coerce'  # invalid dates become NaT instead of error
)

# Drop rows where date conversion failed (optional)
customer_returns = customer_returns.dropna(subset=['order_date'])

# Create MultiIndex
customer_returns = customer_returns.set_index(['customer_id', 'order_date']).sort_index()

# Filter returns
returns_only = customer_returns[customer_returns['return_flag'] == True]

# Group by customer and month
monthly_returns = (
    returns_only
    .groupby([
        returns_only.index.get_level_values(0),  # customer_id
        returns_only.index.get_level_values(1).to_period('M')
    ])['order_id']
    .count()
    .rename('total_returns')
)

# Convert PeriodIndex back to timestamp if needed
monthly_returns.index = monthly_returns.index.set_levels(
    [monthly_returns.index.levels[0],
     monthly_returns.index.levels[1].to_timestamp()]
)

print(monthly_returns)
```
