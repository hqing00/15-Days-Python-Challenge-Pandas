## Question 1

What is the average driver earnings per completed UberPool ride with more than two riders between July 1st and September 30th, 2024? This analysis will help isolate trips that meet specific rider thresholds to understand their impact on driver earnings.

```python
# Filter date range
date_filtered = fct_trips[
    (fct_trips['trip_date'] >= '2024-07-01') &
    (fct_trips['trip_date'] <= '2024-09-30')
]

# Filter for UberPool rides with more than 2 riders
pool_filtered = date_filtered[
    (date_filtered['ride_type'] == 'UberPool') &
    (date_filtered['rider_count'] > 2)
]

# Calculate average driver earnings per trip
avg_earnings = pool_filtered['total_earnings'].mean()

print(avg_earnings)
```

## Question 2

For completed UberPool rides between July 1st and September 30th, 2024, derive a new column calculating earnings per mile (total_earnings divided by total_distance) and then compute the average earnings per mile for rides with more than two riders. This calculation will reveal efficiency metrics for driver compensation.

```python
# Filter date range and UberPool rides
pool_filtered = fct_trips[
    (fct_trips['ride_type'] == 'UberPool') &
    (fct_trips['trip_date'] >= '2024-07-01') &
    (fct_trips['trip_date'] <= '2024-09-30')
]

# Calculate earnings per mile
pool_filtered = pool_filtered.copy()
pool_filtered['earnings_per_mile'] = pool_filtered['total_earnings'] / pool_filtered['total_distance']

# Filter for rides with more than 2 riders
pool_filtered_gt2 = pool_filtered[pool_filtered['rider_count'] > 2]

# Calculate average earnings per mile
avg_earnings_per_mile = pool_filtered_gt2['earnings_per_mile'].mean()

print(round(avg_earnings_per_mile, 3))
```

## Question 3

Identify the combination of rider count and total distance that results in the highest average driver earnings per UberPool ride between July 1st and September 30th, 2024. This analysis directly recommends optimal trip combination strategies to maximize driver earnings.

```python
# Filter for UberPool rides in date range
pool_filtered = fct_trips[
    (fct_trips['ride_type'] == 'UberPool') &
    (fct_trips['trip_date'] >= '2024-07-01') &
    (fct_trips['trip_date'] <= '2024-09-30')
]

# Group by rider_count and total_distance, calculate average earnings
grouped = pool_filtered.groupby(['rider_count', 'total_distance'], as_index=False) \
                       .agg(avg_earnings=('total_earnings', 'mean'))

# Find the combination with highest average earnings
best_combo = grouped.loc[grouped['avg_earnings'].idxmax()]

print("Best combination of rider count & distance:")
print(best_combo)
```
