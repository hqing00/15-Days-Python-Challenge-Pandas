## Question 1

Identify and remove any duplicate entries in the dataset to ensure data quality. How many duplicates were found and removed?

## Question 2

After dropping duplicates, aggregate the data to find the average user interaction time for each number of search results displayed per page. What are the average interaction times?

```python
# Remove duplicate
user_engagement_data_cleaned = user_engagement_data.drop_duplicates()

# average interaction times
avg_interaction_by_results = (
    user_engagement_data_cleaned
    .groupby('search_results_displayed', as_index=False)
    .agg(avg_interaction_time=('interaction_time', 'mean'))
    .sort_values('search_results_displayed')
)

avg_interaction_by_results['avg_interaction_time'] = avg_interaction_by_results['avg_interaction_time'].round(2)

# Result
print(avg_interaction_by_results)
```

## Question 3

Sort the aggregated results from Q2 to determine which number of search results per page has the highest average user interaction time. What is the optimal number of search results per page?

```python
# Remove duplicate
user_engagement_data_cleaned = user_engagement_data.drop_duplicates()

# average interaction times
avg_interaction_by_results = (
    user_engagement_data_cleaned
    .groupby('search_results_displayed', as_index=False)
    .agg(avg_interaction_time=('interaction_time', 'mean'))
    .sort_values('search_results_displayed')
)

# Formant nicely
avg_interaction_by_results['avg_interaction_time'] = avg_interaction_by_results['avg_interaction_time'].round(2)

# Sort by average interaction time descending
optimal_result = avg_interaction_by_results.sort_values('avg_interaction_time', ascending = False)

# Optimal number
optimal_number = optimal_result.iloc[0]

# Result
print("Optimal number of search results per page:")
print(f"Search Results Displayed: {optimal_number['search_results_displayed']}")
print(f"Average Interaction Time: {optimal_number['avg_interaction_time']} seconds")
```
